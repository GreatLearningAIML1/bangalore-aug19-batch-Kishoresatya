```python
# Question -  Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
#              Eg Best Position                Points
#              1                              25000
#              2                              7000    
# Answer   -

laliga["Points"] = pd.to_numeric(laliga["Points"])
laliga_Sum_BestPosition = laliga.groupby("BestPosition")["Points"].agg(["sum"])
print(laliga_Sum_BestPosition.sort_values("BestPosition",ascending = True))
```
