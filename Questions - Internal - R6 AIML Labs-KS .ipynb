### Define weight and bias

```
wt = tf.zeros((4,1),dtype = tf.float32,name = "weights")
bias = tf.zeros((1),dtype = tf.float32, name ="bias")

print(wt)
print(bias)
```

## Question 8

### Get prediction
- Define a function to get prediction
- Approach: prediction = (X * W) + b; here is X is features

```
@tf.function
def pred(features, weight , bias):
  print('Features shape is ' , features.shape)
  print('Weights shape is ',weight.shape)
  #prediction = (features * weight) + bias

  prediction = tf.add(tf.matmul(features, weight),bias)

  return prediction
```

### Calculate loss
- Calculate loss using predictions
- Define a function to calculate loss
- We are calculating mean squared error

```
@tf.function
def loss_cal(actuals , predicted):
  diff = actuals - predicted
  mse = tf.reduce_mean(tf.square(diff))
  return mse
```

## Question 9

### Define a function to train the model
1.   Record all the mathematical steps to calculate Loss
2.   Calculate Gradients of Loss w.r.t weights and bias
3.   Update Weights and Bias based on gradients and learning rate to minimize loss

```
@tf.function
def nn_train(data , target , weight , bias , learning_rate = 0.01):
  
  # Recording operations on tape when calculating loss and prediction
  with tf.GradientTape() as t:
    t.watch([weight,bias])
    prediction = pred(data, weight , bias)
    loss = loss_cal(target , prediction)
  
  # Calculate gradients for loss with respect to weights and bias.
  dw , db = t.gradient(loss,[weight ,bias])

  # update weights and bias
  weight = weight - learning_rate *dw
  bias = bias - learning_rate * db

  return weight, bias
```




## Question 10

### Train the model for 100 epochs 
- Observe the training loss at every iteration

```

for i in range(100):
  weight , bias = nn_train(X_np , y_np , wt , bias , 0.01)
  print("Training loss on iteration " , i ," is " , loss_cal(X_np , pred(X_np , wt , bias)).numpy())
```
