### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
hr.describe()
q1 = hr["Age"].quantile(0.25)  ### lower quartile
q3 = hr["Age"].quantile(0.75)  ### upper quartile
iqr = q3-q1
print(q1)
print(q3)
print(iqr)
```
